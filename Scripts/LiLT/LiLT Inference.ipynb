## Load LiLT Model and Tokenizer

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification, AutoFeatureExtractor, LayoutLMv2Processor

modeName = "pierreguillou/lilt-xlm-roberta-base-finetuned-with-DocLayNet-base-at-linelevel-ml384"

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(modeName)
model = AutoModelForTokenClassification.from_pretrained(modeName)
#feature_extractor = AutoFeatureExtractor.from_pretrained(modeName)

processor = LayoutLMv2Processor.from_pretrained(modeName)

```

```python
from pdf2image import convert_from_path

fileName = "text-and-images.pdf"

# Convert PDF to images
pages = convert_from_path(fileName, dpi=300)

# Save pages as images
for i, page in enumerate(pages):
    page.save(f"./pages/page_{fileName}_{i + 1}.png", "PNG")
```

## OCR Text Extract

```python
import pytesseract
from PIL import Image

# Extract text and bounding boxes
for i, page_image in enumerate(pages):
    ocr_data = pytesseract.image_to_data(page_image, output_type=pytesseract.Output.DICT)
    print(ocr_data)  # Contains text, bounding boxes, and other metadata
```

## Process input for LiLT

```python
from transformers import LayoutLMv2Processor  # Works for LiLT too

processor = LayoutLMv2Processor.from_pretrained(modeName)

# Example data
image = Image.open("./pages/form1.png")
ocr_words = ocr_data["text"]
ocr_boxes = ocr_data["left"], ocr_data["top"], ocr_data["width"], ocr_data["height"]

# Process the data
inputs = processor(image, ocr_words, boxes=ocr_boxes, return_tensors="pt")
```
